```python
from pydantic import BaseModel, EmailStr, PositiveInt, ValidationError
```

```python
class Person(BaseModel):
    name: str
    email: EmailStr
    age: PositiveInt

import json

json_data = {
    
    "age": -30,
    "email": "john_example.com"
}

json_string = json.dumps(json_data)
person = Person.model_validate_json(json_string)
```
